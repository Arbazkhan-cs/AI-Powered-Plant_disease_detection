# Importing Neccessary Librarys

```python
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dropout
from keras.models import Sequential
from keras.losses import SparseCategoricalCrossentropy
from keras.optimizers import Adam
from keras.utils import image_dataset_from_directory
from keras.layers.experimental.preprocessing import Rescaling

import numpy as np
import matplotlib.pyplot as plt
```

# Load Dataset

```python
train_gen = image_dataset_from_directory(directory="/content/New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)/train"
                                              , image_size=(256, 256))

test_gen = image_dataset_from_directory(directory="/content/New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)/valid"
                                              , image_size=(256, 256))
```

# List Diseases

```python
train_gen.class_names
```

```python
classes_len = len(train_gen.class_names)
classes = train_gen.class_names
```

# Rescaling

```python
rescale = Rescaling(scale=1.0/255)
train_gen = train_gen.map(lambda image,label : (rescale(image), label))
test_gen = test_gen.map(lambda image, label: (rescale(image), label))
```

```python
print("Train Generator = ", train_gen)
print("Shape = ", train_gen.element_spec[0])
```

# Model Building

```python
model = Sequential([
    Conv2D(32, (3, 3), activation="relu", padding="same", input_shape=(256, 256, 3)),
    MaxPooling2D(3, 3),

    Conv2D(64, (3, 3), activation="relu", padding="same"),
    MaxPooling2D(4,4),

    Flatten(),

    Dense(1568, activation="relu"),
    Dropout(0.5),

    Dense(classes, activation="softmax")
])
```

```python
model.summary()
```

## Model Compile

```python
model.compile(
    loss = SparseCategoricalCrossentropy(),
    optimizer = Adam(),
    metrics=["accuracy"]
)
```

## Train Model

```python
history = model.fit(train_gen, validation_data=test_gen, epochs=10)
```

# Model Prediction

```python
def predictImage(image):
  # PIL images into NumPy arrays
  numpydata = asarray(image)
  imageNumpy = numpydata.reshape(1, 256, 256, 3)
  predict = model.predict(imageNumpy)
  plt.imshow(numpydata)
  plt.show()
  return classes[np.argmax(predict)]
```

# Visualizing the model history

```python
loss = history.history['loss']
accuracy = history.history['accuracy']
val_loss = history.history['val_loss']
val_accuracy = history.history['val_accuracy']
```

```python
plt.plot(np.arange(0, 10), loss, color="blue", label="loss")
plt.plot(np.arange(0, 10), val_loss, color="red", label="validation loss")
plt.xlabel("Epochs")
plt.ylabel("loss")
plt.title("Loss vs Validation Loss")
plt.legend()
plt.grid(True)
```

```python
plt.plot(np.arange(0, 10), accuracy, color="blue", label="accuracy")
plt.plot(np.arange(0, 10), val_accuracy, color="red", label="validation accuracy")
plt.xlabel("Epochs")
plt.ylabel("accuracy")
plt.title("Accuracy vs Validation Accuracy")
plt.legend()
plt.grid(True)
```
